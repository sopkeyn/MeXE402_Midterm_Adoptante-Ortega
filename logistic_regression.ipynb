## Part 1 - Data Preprocessing

### Importing the dataset

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('heartdisease-UCI.csv')
```

```python
dataset
```

### Data Cleaning

```python
dataset = dataset.drop_duplicates()

dataset
```

```python
dataset['target'] = np.where(dataset['target'] <= 1, 0, dataset['target'])
dataset['target'] = np.where(dataset['target'] > 0, 1, dataset['target'])

dataset
```

### Getting the Inputs and Outputs of the System

```python
X = dataset.iloc[:, :-1].values

X
```

```python
Y = dataset.iloc[:, -1].values

Y
```

### Creating the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_test, Y_train = train_test_split(X, Y, test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
Y_test
```

```python
Y_train
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
```

## Part 2 - Building and training the model

### Building the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
```

### Training the model

```python
model.fit(X_train, Y_train)
```

### Inference

Making the predictons of the data points in the test set

```python
y_pred = model.predict(sc.transform(X_test))
```

```python
y_pred
```

Making the prediction of a single data point with:

## Part 3: Evaluating the model 

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
